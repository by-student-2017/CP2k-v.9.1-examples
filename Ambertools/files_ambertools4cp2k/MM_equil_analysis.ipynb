```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Temperature

```python
time, temp = np.loadtxt("summary.TEMP", unpack=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(time, temp, '-o', linewidth=2)
plt.axvline(30, c="grey", alpha=0.3)
plt.axhline(298, c="grey", alpha=0.3)
plt.title("Temperature", fontsize=18)
plt.xlabel("Simulation time (ps)")
plt.ylabel("Temperature (K)")
plt.text(5,303,"298K", fontsize=12)
plt.ylim(0,350)
plt.xlim(0,60)
plt.savefig("temperature.png")
```

## Energy

```python
time, Etot = np.loadtxt("summary.ETOT", unpack=True)
time, EK = np.loadtxt("summary.EKTOT", unpack=True)
time, EP = np.loadtxt("summary.EPTOT", unpack=True)
```

```python
fig, axs = plt.subplots(3, sharex=True, figsize=(10,6))
fig.suptitle('Total, Kinetic and Potential Energy', fontsize=18)

axs[0].set_title("Total Energy")
axs[0].plot(time, Etot, '-ro',linewidth=2, label="Total energy")
axs[1].set_title("Kinetic Energy")
axs[1].plot(time, EK, '-bo',linewidth=2, label="Kinetic energy")
axs[2].set_title("Potential Energy")
axs[2].plot(time, EP, '-go',linewidth=2, label="Potential energy")
axs[2].set_xlabel("Simulation time (ps)")
axs[0].set_ylabel("Energy (kcal/mol)")
axs[1].set_ylabel("Energy (kcal/mol)")
axs[2].set_ylabel("Energy (kcal/mol)")
plt.xlim(0,60)

axs[0].axvline(30, c="grey", alpha=0.3)
axs[1].axvline(30, c="grey", alpha=0.3)
axs[2].axvline(30, c="grey", alpha=0.3)
plt.savefig("energy.png")
```

## Pressure

```python
time, P = np.loadtxt("summary.PRES", unpack=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(time, P, '-o', linewidth=2)
plt.axhline(1.01325, c="grey", alpha=0.3)
plt.title("Pressure", fontsize=18)
plt.xlabel("Simulation time (ps)")
plt.ylabel("Pressure (bar)")
plt.ylim(-1000,1000)
plt.axvline(30, c="grey", alpha=0.3)
plt.savefig("pressure.png")
```

## Density

```python
time, dens = np.loadtxt("summary.DENSITY", unpack=True)
time, volume = np.loadtxt("summary.VOLUME", unpack=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(time, dens, '-o', linewidth=2)
plt.title("Density", fontsize=18)
plt.xlabel("Simulation time (ps)")
plt.ylabel("Density (g/cm-3)")
plt.ylim(0.9,1.1)
#plt.xlim(0,30)
plt.axhline(1., c="grey", alpha=0.3)
plt.savefig("density.png")
```

```python
fig, axs = plt.subplots(2, sharex=True, figsize=(10,6))
fig.suptitle('Density and Volume of the Simulation Box', fontsize=18)

axs[0].set_title("Density")
axs[0].plot(time, dens, '-ro',linewidth=2, label="Total energy")
axs[1].set_title("Volume")
axs[1].plot(time, volume, '-bo',linewidth=2, label="Kinetic energy")
axs[1].set_xlabel("Simulation time (ps)")
axs[0].set_ylabel("Density (g/cm-3)")
axs[1].set_ylabel("Volume (Angs^3)")
plt.xlim(0,60)

axs[0].axvline(30, c="grey", alpha=0.3)
axs[1].axvline(30, c="grey", alpha=0.3)
plt.savefig("density.png")
```
